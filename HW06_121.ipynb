### Import data and creating the vectorized Train/Test Split 

```python
#Code Block 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Code Block 2


#URL = 'https://data6310.file.core.windows.net/data6310hw/PodcastFilm.xlsx?sp=rl&st=2020-11-12T21:45:20Z&se=2022-12-13T21:45:00Z&sv=2019-12-12&sig=u0hS%2FvAC9s8pJUEYHbt7chCxh09M8z3rGVdw3ifBQKc%3D&sr=f'

podcast_film =  pd.ExcelFile('Homework6_Data/PodcastFilm.xlsx') 
print(podcast_film.sheet_names)
df = podcast_film.parse('podcasts', index_col=0)
```

```python
#Code Block 3

df.head()
```

```python
#Code Block 4

def status5(d):
    if d['rating'] == 5:
        return "5-Star"
    else:
        return "4 or less"
df['Status_5'] = df.apply(status5, axis = 1)
```

```python
#Code Block 5

def status1(d):
    if d['rating'] == 1:
        return "1-Star"
    else:
        return "2 or higher"
df['Status_1'] = df.apply(status1, axis = 1)
```

```python
#Code Block 6

df.sample(10, random_state = 42)
```

```python
#Code Block 7

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Code Block 8

y = df['Status_5']
```

```python
#Code Block 9

X = df['content']
```

### Train test split


```python
#Code Block 10

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Code Block 11

display(y_train.shape)
y_test.shape
```

```python
#Code Block 12

count_vectorizer = CountVectorizer(stop_words='english')
```

```python
#Code Block 13

X_train.head()
```

```python
#Code Block 14

count_train = count_vectorizer.fit_transform(X_train.values)
```

```python
#Code Block 15

count_test = count_vectorizer.transform(X_test.values)
```

```python
#Code Block 16

count_test.shape
```

<h2 style="color:blue;"> Sentiment Analysis for Predicting 5-Star Reviews   </h2>

```python
#Code Block 17

from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
#Code Block 18

nb_classifier_status = MultinomialNB()
```

```python
#Code Block 19

nb_classifier_status.fit(count_train, y_train)
```

```python
#Code Block 20

pred_5 = nb_classifier_status.predict(count_test)
```

```python
#Code Block 21

metrics.accuracy_score(y_test, pred_5)
```

```python
#Code Block 22

pred_5 = pd.DataFrame(pred_5, columns = ['pred_5'])
pred_5.info()
```

```python
#Code Block 23

y_tab = pd.DataFrame(y_test).copy()
y_tab = y_tab.reset_index()
result = pd.concat([y_tab, pred_5], axis=1, join='inner')
result.head(25)
```

### Actual 5-Star Reviews¶

```python
#Code Block 24

result['Status_5'].value_counts()
```

### Predicted 5-Star Reviews¶

```python
#Code Block 25

result['pred_5'].value_counts()
```

```python
#Code Block 26

ct_status_5 = pd.crosstab(result['Status_5'], result['pred_5'])
ct_status_5['Actual Total'] = ct_status_5['4 or less'] + ct_status_5['5-Star']
ct_status_5
```

```python
#Code Block 27

df_sum = pd.DataFrame(ct_status_5.sum(axis=0)).T
df_sum
```

```python
#Code Block 28

ct_status_5 = pd.concat([ct_status_5, df_sum], axis=0)
ct_status_5
```

```python
#Code Block 29

ct_status_5 = ct_status_5.rename({0: 'Predicted Total'})
ct_status_5
```

```python
#Code Block 30

ct_status_5['4orLess_Percent'] = round(ct_status_5['4 or less'] / ct_status_5['Actual Total'], 2)
ct_status_5['5-Star_Percent'] = round(ct_status_5['5-Star'] / ct_status_5['Actual Total'], 2)
ct_status_5
```

### Analysis of 1-Star Crosstab

- We were able to predict 66% of the reviews (4orLess_Percent column - first row) of the 4 or less
- But, we were only able to predict 96% of the 5-Star reviews (5-Star_Percent column - second row)

### Crosstab of Predicted and Actual
- Rows represent actual 1-Star or 2 or higher.
- Columns are predicted 1-Star or 2 or higher.

```python
#Code Block 44

ct_status_1 = pd.crosstab(result_all['Status_1'], result_all['pred_1'])
ct_status_1['Actual Total'] = ct_status_1['2 or higher'] + ct_status_1['1-Star']
ct_status_1
```

```python
#Code Block 45

df_sum = pd.DataFrame(ct_status_1.sum(axis=0)).T
df_sum
```

```python
#Code Block 46

ct_status_1 = pd.concat([ct_status_1, df_sum], axis=0)
ct_status_1
```

```python
#Code Block 47

ct_status_1 = ct_status_1.rename({0: 'Predicted Total'})
ct_status_1
```

```python
#Code Block 48

ct_status_1['1-Star_Percent'] = round(ct_status_1['1-Star'] / ct_status_1['Actual Total'], 2)
ct_status_1['2orHigher_Percent'] = round(ct_status_1['2 or higher'] / ct_status_1['Actual Total'], 2)
ct_status_1
```

### Analysis of 1-Star Crosstab

- We were able to predict 98% of the reviews (2orHigher_Percent column - second row) of the 2 or higher reviews
- But, we were only able to predict 48% of the 1-Star reviews (1-Star_Percent column - first row)
