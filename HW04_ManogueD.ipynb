## 6. Create a Recommender System using PCA

### Section 6a: Create a pivot table named df_comedies_pivot

#### **OBJECTIVE**: To create the first recommender system, we will first run PCA to identify the number of components to include when creating our correlation matrix. (This is similar to HW03, but first we need to create a pivot table version of the users and movies similar to Figure 8.)

```python
df_comedies_pivot = pd.pivot_table(df_comedies_rec,index='title',columns='userId',values='rating')
```

```python
df_comedies_pivot =df_comedies_pivot.fillna(0)
df_comedies_pivot.head()
```

```python
comedy_names = df_comedies_pivot.index
comedy_names
```

```python
# In this section I created a picot table named df_comedies_pivot
```

### Section 6b: Create a PCA for df_comedy_pivot

#### **OBJECTIVE**: Time to reduce the number of dimensions from 243,237 (number of columns in df_comedy_pivot) to a more reasonable number that still captures the explained variance.

```python
from sklearn.decomposition import PCA

n_pcs = 85
model = PCA(n_components=n_pcs)

# Fit the model - return projected data
df_comedy_pca = model.fit_transform(df_comedies_pivot)

#Create a DataFrame of the projected data
df_comedy_pca = pd.DataFrame(df_comedy_pca)
df_comedy_pca.head()
```

```python
names = []
for x in range(n_pcs):
    names.append('PC%d'%(x+1))

names
```

```python
df_comedy_pca.columns = names
df_comedy_pca
```

```python
explained_ratio = model.explained_variance_ratio_
explained_ratio = pd.DataFrame(explained_ratio).reset_index()
explained_ratio = explained_ratio.rename(columns = {0: 'explained_ratio', 'index': 'PC'})
explained_ratio['explained_ratio_sum']= np.cumsum(explained_ratio.explained_ratio)
explained_ratio
```

```python
# In this section I created a PCA for df_comedy_pivot using n_pcs of 5 and then creates a datafram eto show the explained_ratio & explained_ratio_sum
# this explains 70% of variance and a gerneal rule form DR.A is to use that much so you don not lose too much of the variance explained
```

### Section 6c: Create a Correlation Matrix for a Recommender System

```python


corr_mat_comedy = np.corrcoef(df_comedy_pca)
corr_mat_comedy  = pd.DataFrame(corr_mat_comedy, columns=(comedy_names))
corr_mat_comedy.index = comedy_names
corr_mat_comedy
```

```python
corr_ToyStory= pd.DataFrame(corr_mat_comedy['Toy Story'])
corr_ToyStory.shape
```

```python
corr_ToyStory.sort_values(ascending=False, by='Toy Story').head(10)
```

```python
# In this section I created a correlation Matrix for a recommendar system then created a new dataframe for toy story and and sorted the values with the largest correlation to toy story
```

## 7. Create a Recommender System using SVD

### Section 7a: Create a correlation matrix from SVD

#### **OBJECTIVE**: For some reason, I just do not believe the results from the PCA. Why would this be the case? If you go back to the Scenarios/Presentations, you will see that one method for Recommender Systems does a better job than the other when the data is sparse. (Sparse refers to data where most of the values are NaNs. Since most users only rate a few movies, most of the original data in corr_comedy_pivot (prior to filling in NaNs with 0) were NaN. Therefore, we should look at SVD as a better option for a recommender system.)

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=85)
resultant_matrix_SVD = SVD.fit_transform(df_comedies_pivot)
resultant_matrix_SVD.shape
```

```python
df_resultant_matrix_SVD = pd.DataFrame(resultant_matrix_SVD)
df_resultant_matrix_SVD.head()
```

```python
corr_mat_SVD_Comedy = np.corrcoef(resultant_matrix_SVD)
corr_mat_SVD_Comedy = pd.DataFrame(corr_mat_SVD_Comedy, columns=(comedy_names))
corr_mat_SVD_Comedy.index = comedy_names
corr_mat_SVD_Comedy
```

```python
corr_SVD_ToyStory= pd.DataFrame(corr_mat_SVD_Comedy['Click'])
corr_SVD_ToyStory.shape
```

```python
corr_SVD_ToyStory.sort_values(ascending=False, by='Toy Story').head(10)
```
