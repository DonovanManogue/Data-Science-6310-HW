### Clean and Remove Duplicates Data



```python
#Code Block 5


from nltk.corpus import stopwords
nltk.download('stopwords')

from nltk.util import ngrams
from collections import Counter

stop_words = set(stopwords.words('english')) 
```

```python
#Code Block 6

df[df['content'].str.contains("HA - HA - HA - HA")]
```

### To get rid of ha ha ha ha from our text

```python
#Code Block 7

stop_words.add('ha')
```

```python
#Code Block 8

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([wordnet_lemmatizer.lemmatize(word) for word in tokens if word not in stop_words])
    return text

df['cleaned_Reviews'] = df['content'].apply(lambda x: clean_text(x))
df.head()
```

```python
#Code Block 29

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([wordnet_lemmatizer.lemmatize(word) for word in tokens if word not in stop_words])
    return text

df_5_sample['cleaned_Reviews'] = df_5_sample['content'].apply(lambda x: clean_text(x))
df_5_sample.head(2)
```
